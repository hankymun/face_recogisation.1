```python
import os
import cv2
import numpy as np
import gradio as gr
from keras.models import load_model
from google.colab import drive

# 1. Mount Drive để lấy danh sách tên các folder (Tên người)
drive.mount('/content/drive')
DATA_PATH = "/content/drive/MyDrive/60 tấm selfies - sáng T2"
class_names = sorted([f for f in os.listdir(DATA_PATH) if os.path.isdir(os.path.join(DATA_PATH, f))])

# 2. Tải Model (.h5) từ đường dẫn của bạn
MODEL_PATH = "/content/face_recognisation_model.h5"
model = load_model(MODEL_PATH)
IMG_SIZE = 64

# 3. Hàm xử lý logic - Chỉ trả về 1 người duy nhất có điểm cao nhất
def predict_face(image):
    if image is None:
        return "Vui lòng cung cấp hình ảnh!"

    # Tiền xử lý ảnh đầu vào
    img = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
    img = img.astype('float32') / 255.
    img_input = np.expand_dims(img, axis=0)

    # Dự đoán
    predictions = model.predict(img_input)[0]

    # Tìm chỉ số (index) của người có tỷ lệ phần trăm cao nhất
    predicted_class_idx = np.argmax(predictions)

    # Lấy ra tên và độ tự tin tương ứng
    ten_nguoi = class_names[predicted_class_idx]
    ty_le_chinh_xac = predictions[predicted_class_idx] * 100

    # Trả về 1 chuỗi văn bản duy nhất
    return f"👤 Tên: {ten_nguoi} \n🎯"

# 4. Thiết kế giao diện App (Đổi thành gr.Textbox để hiện 1 người)
app = gr.Interface(
    fn=predict_face,
    inputs=gr.Image(label="Chụp hoặc Tải ảnh khuôn mặt lên"),
    outputs=gr.Textbox(label="Kết quả nhận diện"), # Hiển thị dạng chữ gọn gàng
    title="Ứng Dụng Nhận Diện Khuôn Mặt (CNN Model)",
    description="Hệ thống sẽ phân tích và chỉ xuất ra who you are",
    theme="soft"
)

# 5. Kích hoạt chạy App
app.launch(share=True, debug=True)
```
